```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./players_stats_by_season_full_details.csv')
if df["FGM"].isna().values.any():
    mean=df["FGM"].mean()
    df["FGM"].fillna(mean,inplace=True)
```

```python
df["FGM"]
df["SuccessPercentage"]=df["FGM"]/df["FGA"]


```

```python
df["ZScore"]=(df["SuccessPercentage"]-df["SuccessPercentage"].mean())/df["SuccessPercentage"].std()
```

```python
maxZScore=df['ZScore'].max()
conditions = [
    (df['ZScore'] <= maxZScore*0.2),
    (df['ZScore'] > maxZScore*0.2) & (df['ZScore'] <= maxZScore*0.4),
    (df['ZScore'] > maxZScore*0.4) & (df['ZScore'] <= maxZScore*0.6),
    (df['ZScore'] > maxZScore*0.6)
    ]
values = ['Poor', 'Decent', 'Good', 'Excellent']
df['Player Type']=np.select(conditions,values)
df.head(10)
```
